```python
import pandas as pd 
import random
import plotly.graph_objects as go
import json 
from sklearn.metrics import accuracy_score,precision_score,recall_score
import os 
```

```python
def get_label_ls(data_df):
    actual_label_ls=[]
    #print(data_df['actual_label'])

    for ls in data_df['actual_label'].tolist():
        actual_label_ls.extend(json.loads(ls))
        
    pred_label_ls=[]

    for ls in data_df['pred_label'].tolist():
        pred_label_ls.extend(json.loads(ls))
        
    return actual_label_ls,pred_label_ls

def get_label_ls_v2(data_df):
    actual_label_ls=[]

    for ls in data_df['actual_label'].tolist():
        actual_label_ls.append(json.loads(ls))
        
    pred_label_ls=[]

    for ls in data_df['pred_label'].tolist():
        pred_label_ls.append(json.loads(ls))
        
    return actual_label_ls,pred_label_ls


def generate_random_array(length):
    return [random.choice([0, 1]) for _ in range(length)]

def get_label_ls_random(data_df):
    actual_label_ls=[]
    #print(data_df['actual_label'])

    for ls in data_df['actual_label'].tolist():
        actual_label_ls.extend(json.loads(ls))
        
        
    rand_label_ls= generate_random_array(len(actual_label_ls))
        
    return actual_label_ls,rand_label_ls

def get_label_ls_random_v2(data_df):
    actual_label_ls=[]
    #print(data_df['actual_label'])
    rand_label_ls=[]
    for ls in data_df['actual_label'].tolist():
        act_ls=json.loads(ls)
        actual_label_ls.append(act_ls)
        rand_label_ls.append(generate_random_array(len(act_ls)))
        
    return actual_label_ls,rand_label_ls


def get_metrice(pred_ls,test_ls):
    
    accuracy = accuracy_score(test_ls,pred_ls)
    recall = recall_score(test_ls,pred_ls)
    precision = recall_score(test_ls,pred_ls)

    label_ls=[]

    for p,t in zip(pred_ls,test_ls):
        if p == 1 and t == 1:
            label_ls.append('tp')
        elif p == 0 and t == 0:
            label_ls.append('tn')
        elif p == 1 and t == 0:
            label_ls.append('fp')
        elif p == 0 and t == 1:
            label_ls.append('fn')
        
    r_tp = len([i for i in label_ls if i == 'tp']) / len(label_ls)
    r_tn = len([i for i in label_ls if i == 'tn']) / len(label_ls)
    r_fp = len([i for i in label_ls if i == 'fp']) / len(label_ls)
    r_fn = len([i for i in label_ls if i == 'fn']) / len(label_ls)
    
    no_fp = len([i for i in label_ls if i == 'fp'])
    no_fn = len([i for i in label_ls if i == 'fn'])
    
    
    True_label_pred=len([i for i in pred_ls if i >0])/len(pred_ls)
    True_label_act=len([i for i in test_ls if i >0])/len(test_ls)

    result_dict={'accuracy':accuracy,'recall':recall, 'precision':precision,'tp':r_tp,'tn':r_tn, 
                'fp':r_fp, 'fn':r_fn,'n_fp':no_fp, 'n_fn':no_fn,'postive_label_pred':True_label_pred,
                'postive_label_act':True_label_act}
    
    return result_dict


```

```python
opt_files_raw = sorted([f for f in os.listdir('./') if f.endswith('.csv')])
opt_files_name = ['_'.join(f.split('_')[0:]) for f in opt_files_raw]

accur_ls=[]
recall_ls=[]
precision_ls=[]

accur_ls_random=[]
recall_ls_random=[]


fp_ls=[]
fn_ls=[]

#nfp_ls=[]
#nfn1_ls=[]

nfn_ls=[]
no_fault_ls=[]
no_fault1_ls=[]

rnfn_ls=[]
rno_fault_ls=[]
rno_fault1_ls=[]

positive_label_ls=[]

for csv in opt_files_raw:
    data = pd.read_csv(csv)
    act_lab,pred_lab=get_label_ls(data)
    _,rand_label_ls=get_label_ls_random(data)
    
    metrice_dict = get_metrice(pred_lab,act_lab)
    metrice_dict_random = get_metrice(rand_label_ls,act_lab)
    
    accur_ls.append(metrice_dict.get('accuracy'))
    precision_ls.append(metrice_dict.get('precision'))
    fp_ls.append(metrice_dict.get('fp'))
    fn_ls.append(metrice_dict.get('fn'))
    positive_label_ls.append(metrice_dict.get('postive_label_act'))
    recall_ls.append(metrice_dict.get('recall'))

    accur_ls_random.append(metrice_dict_random.get('accuracy'))
    recall_ls_random.append(metrice_dict_random.get('recall'))

    rx_act_lab,rx_pred_lab=get_label_ls_v2(data)
    _,rx_rand_label_ls=get_label_ls_random_v2(data)

    xfn_ls=[]
    xfp_ls=[]
    xallfault_ls=[]
    
    rxfn_ls=[]
    rxfp_ls=[]
    rxallfault_ls=[]
    
    for act_ls,pred_ls,ram_ls in zip(rx_act_lab,rx_pred_lab,rx_rand_label_ls):

        metrice_dict = get_metrice(pred_ls,act_ls)
        metrice_dict_ram = get_metrice(ram_ls,act_ls)

        fn=metrice_dict.get('n_fn')
        fp=metrice_dict.get('n_fp')
        xfn_ls.append(fn)
        xfp_ls.append(fp)
        xallfault_ls.append(fn+fp)
        
        
        rfn=metrice_dict_ram.get('n_fn')
        rfp=metrice_dict_ram.get('n_fp')
        rxfn_ls.append(rfn)
        rxfp_ls.append(rfp)
        rxallfault_ls.append(rfn+rfp)
        
    
    #nfp_ls.append(len([i for i in xfp_ls if i == 0])/len(xfp_ls))
    #nfn1_ls.append(len([i for i in xfn_ls if i <= 1])/len(xfn_ls))
    #'n_fp':nfp_ls, 'n_fn':nfn_ls,  'n_fn1':nfn1_ls 
    nfn_ls.append(len([i for i in xfn_ls if i == 0])/len(xfn_ls))

    no_fault_ls.append(len([i for i in xallfault_ls if i == 0])/len(xallfault_ls))
    no_fault1_ls.append(len([i for i in xallfault_ls if i <= 1])/len(xallfault_ls))
    
    rnfn_ls.append(len([i for i in rxfn_ls if i == 0])/len(rxfn_ls))
    rno_fault_ls.append(len([i for i in rxallfault_ls if i == 0])/len(rxallfault_ls))
    rno_fault1_ls.append(len([i for i in rxallfault_ls if i <= 1])/len(rxallfault_ls))
    
    
    
    
result=pd.DataFrame({'file':opt_files_name, 'accuracy':accur_ls, 'precision':precision_ls,'fp':fp_ls, 'fn':fn_ls, 
              'no_fault':no_fault_ls,'no_fault1':no_fault1_ls, 'n_fn':nfn_ls,  
              'postive_label_act':positive_label_ls,'recall':recall_ls,
              'no_fault_ram':rno_fault_ls,'no_fault1_ram':rno_fault1_ls, 'n_fn_ram':rnfn_ls })


result
```

```python

def spl_csv(filename):
    
    just_filename=filename[:-4]
    
    data = pd.read_csv(filename)
    half_point=int((max(data['idx'].tolist())+1)/2)

    sub_data_1=data[data['idx']<half_point]
    sub_data_2= data[data['idx']>=half_point]
    
    sub_data_1.to_csv(just_filename+'_a.csv',index=True)
    sub_data_2.to_csv(just_filename+'_b.csv',index=True)
    
    
```

```python
spl_csv('typA_200_05012024_all_label.csv')
```
